```python
from pylab import *

# make a square figure and axes
figure(1, figsize=(6,6))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.
labels = 'UK-based', 'Non-UK', 'Unidentified location'
fracs = [54, 8.5, 37.5]
explode=(0, 0, 0)

pie(fracs, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

title('User location for tweets about Brexit and migration', bbox={'facecolor':'0.8', 'pad':5})

show()
```
